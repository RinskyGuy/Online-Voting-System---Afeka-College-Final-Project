## **Check Thresholds**

<a id="sub-section-three-one"></a>
### **IDs**

```python
true_match = pd.read_csv(os.path.join(data_dir, 'id_ocr_true_distances.csv')).drop(columns=['time', 'id'])
false_match = pd.read_csv(os.path.join(data_dir, 'id_ocr_false_distances.csv')).drop(columns=['time', 'id'])

plot(true_match, false_match)
```

```python
dist_field = {'name':'avg_text_distance'}
metrics = get_metrics(true_match, false_match, [dist_field])
text_thresholds = get_best_thresholds(metrics, ['accuracy', 'f1_score'], n=5)
text_thresholds
```

```python
cosine_dist_field = {'name':'cosine_distance'}
metrics = get_metrics(true_match, false_match, [cosine_dist_field])
euclidean_thresholds = get_best_thresholds(metrics, ['accuracy', 'f1_score'], n=5)
euclidean_thresholds
```
